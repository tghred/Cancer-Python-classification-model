```python
def check_func(cols,axis=1):
    
    status=cols[0]
    for i in status:
        if i == 'M':
            return status
        elif i =='S':
            return status
        else:
            return 'M'
    
```

```python
book1['status']=book1['status'].apply(func=check_func,axis=1)
```
